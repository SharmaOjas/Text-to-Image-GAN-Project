```python
from IPython import get_ipython
from IPython.display import display
# %%
!pip install sentence-transformers
!pip install nltk
# %%
import nltk
nltk.download('punkt')
nltk.download('stopwords')
# Add this line to download the punkt_tab resource
nltk.download('punkt_tab')
# %%
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return tokens

# Example usage:
text_example = "A vibrant, large bird with red and blue feathers is flying high in the sky."
processed_tokens = preprocess_text(text_example)
print(f"Original text: {text_example}")
print(f"Processed tokens: {processed_tokens}")
```

```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained Sentence-BERT model
model = SentenceTransformer('all-mpnet-base-v2')

# Assuming you have the 'processed_tokens' from the previous step
processed_text = " ".join(processed_tokens) # Join tokens back into a string for Sentence-BERT
text_embedding = model.encode(processed_text)

print(f"Shape of the text embedding: {text_embedding.shape}")
print(f"First 10 values of the embedding: {text_embedding[:10]}")
```

```python
import torch.nn as nn
import torch

class Generator(nn.Module):
    def __init__(self, embedding_dim, noise_dim, output_channels, image_size=64):
        super(Generator, self).__init__()
        self.embedding_dim = embedding_dim
        self.noise_dim = noise_dim
        self.image_size = image_size

        self.linear = nn.Linear(embedding_dim + noise_dim, (image_size // 16) * (image_size // 16) * 256)
        self.bn1 = nn.BatchNorm1d((image_size // 16) * (image_size // 16) * 256)
        self.relu = nn.ReLU()

        self.conv_transpose1 = nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1)
        self.bn2 = nn.BatchNorm2d(128)

        self.conv_transpose2 = nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1)
        self.bn3 = nn.BatchNorm2d(64)

        self.conv_transpose3 = nn.ConvTranspose2d(64, output_channels, kernel_size=4, stride=2, padding=1)
        self.tanh = nn.Tanh() # Output image pixels in the range [-1, 1]

    def forward(self, text_embedding, noise):
        # Concatenate text embedding and noise
        x = torch.cat([text_embedding, noise], 1)
        x = self.linear(x)
        # Ensure the input to BatchNorm1d has more than one element in the batch dimension if in training mode
        # For a batch size of 1, BatchNorm1d will throw an error in training mode.
        # During inference (model.eval()), it will use running statistics and won't have this issue.
        # However, for demonstrating the forward pass with a single example,
        # it's better to use a batch size > 1 or set the model to evaluation mode.
        # We will use a batch size > 1 for this example to simulate a real batch.
        x = self.bn1(x)
        x = self.relu(x)
        x = x.view(x.size(0), 256, self.image_size // 16, self.image_size // 16) # Reshape for convolutional layers

        x = self.conv_transpose1(x)
        x = self.bn2(x)
        x = self.relu(x)

        x = self.conv_transpose2(x)
        x = self.bn3(x)
        x = self.relu(x)

        image = self.conv_transpose3(x)
        image = self.tanh(image)
        return image

# Instantiate the Generator
embedding_dim = 768
noise_dim = 100
output_channels = 3 # For RGB images
image_size = 64
generator = Generator(embedding_dim, noise_dim, output_channels, image_size)

# Example of passing input through the generator
# Use a batch size > 1 to avoid the BatchNorm1d error in training mode
batch_size = 4 # Increased batch size
example_embedding = torch.randn(batch_size, embedding_dim)
example_noise = torch.randn(batch_size, noise_dim)
generated_image = generator(example_embedding, example_noise)

print(f"Shape of the generated image: {generated_image.shape}")
```

```python
class Discriminator(nn.Module):
    def __init__(self, input_channels, embedding_dim, image_size=16):
        super(Discriminator, self).__init__()
        self.image_size = image_size

        self.conv1 = nn.Conv2d(input_channels, 64, kernel_size=4, stride=2, padding=1)
        self.leaky_relu = nn.LeakyReLU(0.2)

        self.conv2 = nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1)
        self.bn1 = nn.BatchNorm2d(128)

        # The output of the convolutional layers will be flattened and combined with the text embedding
        self.fc_image = nn.Linear(128 * (image_size // 4) * (image_size // 4), 512) # For 16x16 image, this is 128 * 4 * 4
        self.fc_text = nn.Linear(embedding_dim, 512)
        self.fc_combined = nn.Linear(1024, 1) # Combine image and text features
        self.sigmoid = nn.Sigmoid()

    def forward(self, image, text_embedding):
        x_image = self.conv1(image)
        x_image = self.leaky_relu(x_image)

        x_image = self.conv2(x_image)
        x_image = self.bn1(x_image)
        x_image = self.leaky_relu(x_image)

        x_image = x_image.view(x_image.size(0), -1) # Flatten

        image_features = self.fc_image(x_image)
        text_features = self.fc_text(text_embedding)

        combined_features = torch.cat([image_features, text_features], 1)
        output = self.sigmoid(self.fc_combined(combined_features))
        return output

# Instantiate the Discriminator
input_channels = 3
embedding_dim = 768
image_size = 16
discriminator = Discriminator(input_channels, embedding_dim, image_size)

# Example of passing input through the discriminator
example_image = torch.randn(4, input_channels, image_size, image_size)
example_embedding = torch.randn(4, embedding_dim)
discriminator_output = discriminator(example_image, example_embedding)

print(f"Shape of the discriminator output: {discriminator_output.shape}")
```

```python
import torch
import torch.optim as optim
import torch.nn as nn

# Assuming generator and discriminator are already defined
# Assuming you have a way to get real images and text embeddings (dataloader)

# Loss function
criterion = nn.BCELoss()

# Optimizers
lr = 0.0002
optimizer_g = optim.Adam(generator.parameters(), lr=lr, betas=(0.5, 0.999))
optimizer_d = optim.Adam(discriminator.parameters(), lr=lr, betas=(0.5, 0.999))

# Fixed noise for visualization
fixed_noise = torch.randn(64, noise_dim)
# You would also need fixed text embeddings for consistent generation

num_epochs = 10 # Just for illustration

for epoch in range(num_epochs):
    # Simulate a batch of real images and text embeddings
    real_images = torch.randn(64, 3, 16, 16) # Replace with actual data loading
    real_embeddings = torch.randn(64, embedding_dim) # Replace with actual embeddings

    # Labels for real and fake samples
    real_labels = torch.ones((64, 1))
    fake_labels = torch.zeros((64, 1))

    # --- Train Discriminator ---
    optimizer_d.zero_grad()

    # Real batch
    output_real = discriminator(real_images, real_embeddings)
    loss_real = criterion(output_real, real_labels)

    # Fake batch
    noise = torch.randn(64, noise_dim)
    fake_images = generator(real_embeddings, noise) # Using real embeddings to guide generation
    output_fake = discriminator(fake_images.detach(), real_embeddings) # Detach to avoid generator gradients
    loss_fake = criterion(output_fake, fake_labels)

    loss_d = (loss_real + loss_fake) / 2
    loss_d.backward()
    optimizer_d.step()

    # --- Train Generator ---
    optimizer_g.zero_grad()
    # Generate fake images again
    fake_images_gen = generator(real_embeddings, noise)
    output_fake_gen = discriminator(fake_images_gen, real_embeddings)
    loss_g = criterion(output_fake_gen, real_labels) # Try to make the discriminator think they are real

    loss_g.backward()
    optimizer_g.step()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss D: {loss_d.item():.4f}, Loss G: {loss_g.item():.4f}")

# After training, you would use the generator to create images from text embeddings
# with torch.no_grad():
#     generated_images = generator(some_text_embedding, fixed_noise)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Define the set of colors we want to classify
color_labels = ["red", "blue", "yellow"]
num_classes = len(color_labels)

class TextColorClassifier(nn.Module):
    def __init__(self, embedding_dim, num_classes):
        super(TextColorClassifier, self).__init__()
        self.linear1 = nn.Linear(embedding_dim, 128)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(128, num_classes)

    def forward(self, embedding):
        x = self.linear1(embedding)
        x = self.relu(x)
        x = self.linear2(x)
        return x # Output logits (not probabilities yet)

# Instantiate the classifier
embedding_dim = 768
classifier = TextColorClassifier(embedding_dim, num_classes)

# Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(classifier.parameters(), lr=0.001)

# Simulate some data (text descriptions and corresponding color labels)
train_texts = [
    "A bird with bright red wings.",
    "The feathers are mostly blue.",
    "It has a yellow crest on its head.",
    "Red spots on its chest.",
    "A very blue tail.",
    "Yellow underparts.",
    "A bird that is red and black.", # Might be harder to classify as purely one color
    "Sky blue feathers.",
    "Golden yellow beak.",
    "Crimson red throat."
]
train_colors = [0, 1, 2, 0, 1, 2, 0, 1, 2, 0] # 0: red, 1: blue, 2: yellow

# Preprocess texts and get embeddings
from sentence_transformers import SentenceTransformer
pretrained_model = SentenceTransformer('all-mpnet-base-v2')

train_embeddings = pretrained_model.encode(train_texts)
train_embeddings_tensor = torch.tensor(train_embeddings, dtype=torch.float32)
train_labels_tensor = torch.tensor(train_colors, dtype=torch.long)

# Training loop (simplified)
epochs = 100
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = classifier(train_embeddings_tensor)
    loss = criterion(outputs, train_labels_tensor)
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# Evaluation (on the training data for simplicity)
with torch.no_grad():
    classifier.eval()
    predicted_logits = classifier(train_embeddings_tensor)
    _, predicted_labels = torch.max(predicted_logits, 1)
    predicted_labels_np = predicted_labels.numpy()
    true_labels_np = train_labels_tensor.numpy()

    accuracy = np.mean(predicted_labels_np == true_labels_np)
    print(f'Accuracy on training data: {accuracy:.4f}')

    print('\nConfusion Matrix:')
    cm = confusion_matrix(true_labels_np, predicted_labels_np)
    print(cm)

    print('\nClassification Report:')
    print(classification_report(true_labels_np, predicted_labels_np, target_names=color_labels))
```
